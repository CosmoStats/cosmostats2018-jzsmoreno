```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import math 
import cmath
from IPython.display import display
from IPython.core.pylabtools import figsize, getfigs
plt.style.use('classic')
%matplotlib inline
```

$Para$

$P(k)=\frac{1}{\left ( 2\pi\right )^{d} }\int_{0}^{\infty }\xi (r)e^{-i\mathbf{k}\cdot \mathbf{r}}dr$

$donde$ $\xi (r)=\left ( \frac{r}{r_{o}} \right )^{\gamma }$

$Calculamos$ $para$ 1D $donde$ $\mathbf{r}=r\widehat{r}$  $(1)$

$P(k)=\frac{1}{\left ( 2\pi\right ) }\int_{0}^{\infty }\left ( \frac{r}{r_{o}} \right )^{\gamma }e^{-i\mathbf{k}\cdot \mathbf{r}}dr$ (2)

$=\frac{1}{\left ( 2\pi\right ) }\int_{0}^{\infty }\left ( \frac{r}{r_{o}} \right )^{\gamma }e^{-ikrcos\theta }dr$ (3)

$Usando$ (1) en (3)

$P(k)=\frac{1}{\left ( 2\pi\right ) }\int_{0}^{\infty }\left ( \frac{r}{r_{o}} \right )^{\gamma }e^{-ikr }dr$ (4)

$Se$ $propone$ $el$ $siguiente$ $cambio$ $de$ $variable$ $u=ikr$

$donde$ $du=ikdr$, $por$ $lo$ $que$ $de$ (4) $se$ $tiene:$

$P(k)=\frac{1}{2\pi r_{o}^{\gamma }}\int_{0}^{\infty }\frac{u^{\gamma }}{(ik)^{\gamma +1}} e^{-u }dr$ (5)

$Haciendo$ $k_{p}=ir_{o}\left ( \frac{r_{o}}{2\pi } \right )^{\gamma +1}$

$=\left ( \frac{k}{k_{p}} \right )^{-\gamma -1}\int_{0}^{\infty }u^{\gamma } e^{-u }dr$ (6)

$De$ $(6)$ $vemos$ $que$ $la$ $integral$ $puede$ $ser$ $expresada$ $en$ $terminos$ $de$ $la$ $funcion$ $Gamma$

$=\left ( \frac{k}{k_{p}} \right )^{-\gamma -1}\Gamma (\gamma +1)$ 

$Finalmente$

$P(k)=A^{2}\left ( \frac{k}{k_{p}} \right )^{-\gamma -1}$

$con$ $los$ $valores$ $\gamma = 1.77$ and $r_{o}^{\gamma}=5.4$

```python
j=complex(0,1) # Se define el vector complejo, (0,1)
x=np.linspace(10,0.1,100)
y=(x**1.77)/5.4
k=1./x
kp=j*(2.5928)*(14.0015/(2*math.pi)**2.77)
p=2.0*(k/kp)**(-2.77)
v=np.linalg.norm(y)
u=np.linalg.norm(abs(p))
plt.figure(figsize=(7,5))
plt.plot(x,y/v,'-' ,label='$\zeta$ $(r)$')
plt.plot(k,abs(p)/u,'-' ,label='$P$ $(k)$')
plt.xlabel("$r$,$k$",fontsize = 20)   # Establece el título del eje x
plt.ylabel("$\zeta$ $(r)$, $P$ $(k)$",fontsize = 20)   # Establece el título del eje y
plt.legend()
plt.autoscale
#plt.yscale('log')
#plt.xscale('log')
plt.show()
```

$Analogamente$ $para$ 2D

$P(k)=\frac{1}{\left ( 2\pi\right )^{2} }\int_{0}^{2\pi }\int_{0}^{\infty }\left ( \frac{r}{r_{o}} \right )^{\gamma }e^{-i\mathbf{k}\cdot \mathbf{r}}rdrd\theta $

$=\frac{1}{\left ( 2\pi\right )^{2} }\int_{0}^{2\pi }\int_{0}^{\infty }\left ( \frac{r}{r_{o}} \right )^{\gamma }e^{-ikrcos\theta }rdrd\theta $ (1)

$Haciendo$ $el$ $mismo$ $cambio$ $de$ $variable$

$=\int_{0}^{2\pi }\int_{0}^{\infty }\frac{u^{\gamma +1}}{r_{o}^{\gamma }\left ( ikcos\theta  \right )^{\gamma +2} }e^{-u }dud\theta $ (2)

$=\left ( \frac{k}{k_{p}} \right )^{-\gamma -2}\int_{0}^{2\pi }\left (cos\theta  \right )^{-\gamma -2}d\theta \int_{0}^{\infty }u^{\gamma +1}e^{-u }du$ (3)

$donde$ $k_{p}=ir_{o}\left ( \frac{r_{o}}{4\pi ^{2}} \right )^{\gamma +2}$

$De$ (3) $vemos$ $que$ $la$ $ultima$ $integral$ $puede$ $ser$ $expresada$ $en$ $terminos$ $de$ $la$ $funcion$ $Gamma$

$P(k)=\left ( \frac{k}{k_{p}} \right )^{-\gamma -2}\Gamma (\gamma +2)\int_{0}^{2\pi }\left (cos\theta  \right )^{-\gamma -2}d\theta $ (4)

$Expandiendo$ $en$ $serie$

$=\left ( \frac{k}{k_{p}} \right )^{-\gamma -2}\Gamma (\gamma +2)\int_{0}^{2\pi }\left (\sum \frac{(-1)^{n}\theta ^{2n}}{(2n)!}  \right )^{-\gamma -2}d\theta $ (5)

$=\left ( \frac{k}{k_{p}} \right )^{-\gamma -2}\Gamma (\gamma +2)\int_{0}^{2\pi }\left (\sum \frac{((-1)^{n}\theta ^{2n})^{(-\gamma -2)}}{((2n)!)^{-\gamma -2}}  \right )d\theta $ (6)

$Intercambiando$ $la$ $sumatoria$

$P(k)=\left ( \frac{k}{k_{p}} \right )^{-\gamma -2}\Gamma (\gamma +2)\left (\sum\frac{((-1)^{n(-\gamma -2)}}{((2n)!)^{-\gamma -2}}\int_{0}^{2\pi }\theta ^{2n(-\gamma -2)}d\theta \right )$ (7)

$donde$ $\int_{0}^{2\pi }\theta ^{2n(-\gamma -2)}d\theta=\frac{\left ( 2\pi  \right )^{1-2(2+\gamma )n}}{2(2+\gamma )n-1} $

$P(k)=\left ( \frac{k}{k_{p}} \right )^{-\gamma -2}\Gamma (\gamma +2)\left ( \sum\frac{((-1)^{n(-\gamma -2)}}{((2n)!)^{-\gamma -2}}\cdot \frac{\left ( 2\pi  \right )^{1-2(2+\gamma )n}}{2(2+\gamma )n-1}  \right )$ (8)

$Asi$ $pues$ $\left ( \sum\frac{((-1)^{n(-\gamma -2)}}{((2n)!)^{-\gamma -2}}\cdot \frac{\left ( 2\pi  \right )^{1-2(2+\gamma )n}}{2(2+\gamma )n-1}  \right )\sim \sum \frac{2\pi\left ( cos(2\pi ) \right )^{-\gamma -2}}{2(2+\gamma )n-1}$

$De$ $la$ $sumatoria$ $\sum \frac{2\pi\left ( cos(2\pi ) \right )^{-\gamma -2}}{2(2+\gamma )n-1}=2\pi\left ( 1-\frac{1}{2(2+\gamma )-1} -\frac{1}{4(2+\gamma )-1}-...- \right)\sim 2\pi $

$Finalmente$

$P(k)= 2\pi \left( \frac{k}{k_{p}} \right )^{-\gamma -2}\Gamma (\gamma +2)$ (9)

$=A^{2}\left( \frac{k}{k_{p}} \right )^{-\gamma -2}$

```python
def normalize(X):
    NORM = []
    for i in X:
        norm= (i-np.min(X))/(np.max(X)-np.min(X))
        NORM.append(norm)
    return NORM
```

```python
def f(x, y):
    j=complex(0,1) # Se define el vector complejo, (0,1)
    kp=j*(2.5928)*(14.0015/(4*(math.pi**2))**3.77)
    k=np.sqrt(x ** 2 + y ** 2)
    return abs(2*np.pi*6.0*(k/kp)**(-3.77))

def g(x, y):
    r=np.sqrt(x ** 2 + y ** 2)
    return (r**1.77)/5.4

x=np.linspace(.1,1,100)
y=np.linspace(.1,1,100)
k1=1./x
k2=1./y

x, y = np.meshgrid(x, y)
gy = g(x, y)

k1, k2 = np.meshgrid(k1, k1)
p = f(k1, k2)

gy = normalize(gy)
p = normalize(p)

ax = plt.axes(projection='3d')
ax.plot_surface(k1,k2,gy,cmap='viridis', edgecolor='none')
ax.set_xlabel('$r_{x}$',fontsize = 15)
ax.set_ylabel('$r_{y}$',fontsize = 15)
ax.set_zlabel('$\zeta$ $(r)$',fontsize = 15)
ax.view_init(10, 65)
```
