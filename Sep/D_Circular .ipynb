```python
from mpl_toolkits import mplot3d
import numpy as np
import random
import matplotlib.pyplot as plt
import scipy.stats as st
import math 
from IPython.display import display
from IPython.core.pylabtools import figsize, getfigs
plt.style.use('classic')
%matplotlib inline
```

### Generamos una variable aleatoria x, y uniforme  

```python
print("Ingrese cuantos puntos desea generar")
n=int(input())
print("Ingrese el diámetro del circulo")
d=float(input())
lx=np.zeros(n)
ly=np.zeros(n)
for i in range(n):
    x=random.uniform(-d/2.0,d/2.0)
    y=random.uniform(-d/2.0,d/2.0)
    r=math.sqrt(x*x+y*y)
    if(r<=d/2.0):
        lx[i]=x
        ly[i]=y
plt.plot(lx,ly, 'bo')
plt.rcParams["figure.figsize"] = [d,d]
xlim=(-d/2.0,d/2.0)
ylim=(-d/2.0,d/2.0)
plt.show()
```

### Transformación a coordenadas esféricas usando el Jacobiano

$r{}'=r^{2}, \theta {}'=\theta , x=r{}'cos\theta , y=r{}'sin\theta \mid 0\leq r\leq \sqrt{d/2}, -\pi \leq \theta \leq \pi $

```python
print("Ingrese cuantos puntos desea generar")
n=int(input())
print("Ingrese el diámetro del circulo")
d=float(input())
lc=np.zeros(n)
ls=np.zeros(n)
for i in range(n):
    phi=random.uniform(-math.pi,math.pi)
    r=math.sqrt(random.uniform(0.0,(d/2.0)**2.0))
    theta=phi
    lc[i]=r*math.sin(theta)
    ls[i]=r*math.cos(theta)
plt.plot(ls,lc, 'bo')
#plt.rcParams["figure.figsize"] = [d,d]
xlim=(-d/2.0,d/2.0)
ylim=(-d/2.0,d/2.0)
plt.show()
```

### Transformación usando el Jacobiano en coordenadas esféricas

$r{}'=r^{2}, \theta {}'=acos(\theta) , x=r{}'sin\theta cos\varphi , y=r{}'sin \varphi sin\theta, z=r{}'cos\theta \mid 0\leq r\leq \sqrt{d/2}, -1 \leq \theta \leq 1, 0\leq \varphi\leq 2\pi$ 

```python
ax = plt.axes(projection='3d')
print("Ingrese cuantos puntos desea generar")
n=int(input())
print("Ingrese el diámetro del circulo")
d=float(input())
lc=np.zeros(n)
ls=np.zeros(n)
lz=np.zeros(n)
for i in range(n):
    phi=random.uniform(-math.pi,math.pi)
    r=math.sqrt(random.uniform(0.0,(d/2.0)**2))
    theta=math.acos(random.uniform(-1,1))
    lc[i]=r*math.sin(theta)*math.cos(phi)
    ls[i]=r*math.sin(theta)*math.sin(phi)
    lz[i]=r*math.cos(theta)
ax.scatter3D(lc, ls, lz, cmap='viridis', linewidth=0.5);
#plt.rcParams["figure.figsize"] = [d,d]
xlim=(-d/2.0,d/2.0)
ylim=(-d/2.0,d/2.0)
zlim=(-d/2.0,d/2.0)
ax.view_init(20, 45)
plt.show()
```
